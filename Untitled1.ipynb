⭐ 1.Data Import & Preview

```python
import pandas as pd

# Load all necessary CSV files into pandas DataFrames
orders = pd.read_csv('/content/orders.csv')
order_details = pd.read_csv('/content/order_details.csv')
pizzas = pd.read_csv('/content/pizzas.csv')
pizza_types = pd.read_csv('/content/pizza_types.csv')

def get_master_df():
    df = order_details.merge(pizzas, on="pizza_id")
    df = df.merge(pizza_types, on="pizza_type_id")
    df = df.merge(orders, on="order_id")
    df["revenue"] = df["quantity"] * df["price"]
    df["date"] = pd.to_datetime(df["date"])
    df["hour"] = pd.to_datetime(df["time"], format="%H:%M:%S").dt.hour
    return df

master_df = get_master_df()

# Generate the required dataframes from master_df
cumulative_revenue_df = master_df.groupby('date')['revenue'].sum().cumsum().reset_index()
cumulative_revenue_df.columns = ['date', 'cum_rev']

top_3_by_category_df = master_df.groupby(['category', 'name'])['revenue'].sum().reset_index()
top_3_by_category_df = top_3_by_category_df.sort_values(by=['category', 'revenue'], ascending=[True, False])
top_3_by_category_df = top_3_by_category_df.groupby('category').head(3).reset_index(drop=True)
top_3_by_category_df.rename(columns={'revenue': 'cat_pizza_rev'}, inplace=True)

category_quantity_df = master_df.groupby('category')['quantity'].sum().reset_index()

avg_pizzas_per_day = master_df.groupby('date')['quantity'].sum().mean()
avg_pizzas_per_day_df = pd.DataFrame([{'avg_number_per_day': avg_pizzas_per_day}])

top_pizzas_revenue_df = master_df.groupby('name')['revenue'].sum().nlargest(3).reset_index()
top_pizzas_revenue_df.columns = ['pizza_type', 'total_revenue']

hourly_orders_df = master_df.groupby('hour')['order_id'].nunique().reset_index()
hourly_orders_df.columns = ['hr', 'hourly_orders']

top_pizzas_df = master_df.groupby('name')['quantity'].sum().nlargest(5).reset_index()
top_pizzas_df.columns = ['pizza_type', 'total_quantity']

popular_size_df = master_df.groupby('size')['quantity'].sum().reset_index()
popular_size_df.columns = ['size', 'total_quantity']

# Display the head of each DataFrame to confirm loading
print("Cumulative Revenue DataFrame Head:")
display(cumulative_revenue_df.head())

print("\nTop 3 by Category DataFrame Head:")
display(top_3_by_category_df.head())

print("\nCategory Quantity DataFrame Head:")
display(category_quantity_df.head())

print("\nAverage Pizzas per Day DataFrame Head:")
display(avg_pizzas_per_day_df.head())

print("\nTop Pizzas by Revenue DataFrame Head:")
display(top_pizzas_revenue_df.head())

print("\nHourly Orders DataFrame Head:")
display(hourly_orders_df.head())

print("\nTop Pizzas DataFrame Head:")
display(top_pizzas_df.head())

print("\nPopular Size DataFrame Head:")
display(popular_size_df.head())
```

```python
from flask import Flask, jsonify, request
from flask_cors import CORS
import pandas as pd
from pyngrok import ngrok
import threading

app = Flask(__name__)
CORS(app)

# Load your CSV files
orders        = pd.read_csv("orders.csv")
order_details = pd.read_csv("order_details.csv")
pizzas        = pd.read_csv("pizzas.csv")
pizza_types   = pd.read_csv("pizza_types.csv")

def get_master_df():
    df = order_details.merge(pizzas, on="pizza_id")
    df = df.merge(pizza_types, on="pizza_type_id")
    df = df.merge(orders, on="order_id")
    df["revenue"] = df["quantity"] * df["price"]
    df["date"]    = pd.to_datetime(df["date"])
    df["hour"]    = pd.to_datetime(df["time"], format="%H:%M:%S").dt.hour
    return df

@app.route("/")
def home():
    return jsonify({"message": "🍕 Pizza Sales API is running!", "author": "Ishika Kushwah"})

@app.route("/api/summary")
def summary():
    df = get_master_df()
    return jsonify({
        "total_orders"      : int(orders["order_id"].nunique()),
        "total_revenue"     : round(float(df["revenue"].sum()), 2),
        "total_pizzas_sold" : int(df["quantity"].sum()),
        "top_pizza"         : df.groupby("name")["quantity"].sum().idxmax(),
        "avg_order_value"   : round(float(df.groupby("order_id")["revenue"].sum().mean()), 2)
    })

@app.route("/api/pizzas/top")
def top_pizzas():
    df    = get_master_df()
    limit = int(request.args.get("limit", 5))
    top   = df.groupby("name").agg(
                total_quantity=("quantity","sum"),
                total_revenue =("revenue","sum")
            ).sort_values("total_quantity", ascending=False).head(limit).reset_index()
    top["total_revenue"] = top["total_revenue"].round(2)
    return jsonify(top.to_dict(orient="records"))

@app.route("/api/orders/peak-hours")
def peak_hours():
    df     = get_master_df()
    hourly = df.groupby("hour")["order_id"].nunique().reset_index()
    hourly.columns = ["hour", "order_count"]
    peak   = int(hourly.loc[hourly["order_count"].idxmax(), "hour"])
    return jsonify({"peak_hour": f"{peak}:00", "hourly_data": hourly.to_dict(orient="records")})

@app.route("/api/category/revenue")
def category_revenue():
    df  = get_master_df()
    cat = df.groupby("category").agg(
              total_revenue  =("revenue","sum"),
              total_quantity =("quantity","sum")
          ).sort_values("total_revenue", ascending=False).reset_index()
    cat["total_revenue"] = cat["total_revenue"].round(2)
    return jsonify(cat.to_dict(orient="records"))

# Set ngrok authtoken and start Flask + ngrok
# This token was provided by the user in cell yyVdTfPvnfVx
ngrok.set_auth_token("3C8IxJKTIMuUhmYIKgFJRyIMhNV_4MC1gspaS9C7mcUdook6r")
public_url = ngrok.connect(5000)
print(f"\n✅ Your Public API Link: {public_url}")
print(f"📊 Summary: {public_url}/api/summary")
print(f"🍕 Top Pizzas: {public_url}/api/pizzas/top?limit=5")
print(f"⏰ Peak Hours: {public_url}/api/orders/peak-hours")

threading.Thread(target=app.run, kwargs={"port": 5000}).start()
```
