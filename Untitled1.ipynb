```python
import pandas as pd

# Load all necessary CSV files into pandas DataFrames
cumulative_revenue_df = pd.read_csv('cumulative_revenue.csv')
top_3_by_category_df = pd.read_csv('top_3_by_category..csv') # Note: double dot in filename
category_quantity_df = pd.read_csv('category_quantity.csv')
avg_pizzas_per_day_df = pd.read_csv('avg_pizzas_per_day.csv')
top_pizzas_revenue_df = pd.read_csv('top_pizzas_by_revenue.csv')
hourly_orders_df = pd.read_csv('hourly_orders.csv')
top_pizzas_df = pd.read_csv('top_pizzas.csv')
popular_size_df = pd.read_csv('popular_size.csv')

# Display the head of each DataFrame to confirm loading
print("Cumulative Revenue DataFrame Head:")
display(cumulative_revenue_df.head())

print("\nTop 3 by Category DataFrame Head:")
display(top_3_by_category_df.head())

print("\nCategory Quantity DataFrame Head:")
display(category_quantity_df.head())

print("\nAverage Pizzas per Day DataFrame Head:")
display(avg_pizzas_per_day_df.head())

print("\nTop Pizzas by Revenue DataFrame Head:")
display(top_pizzas_revenue_df.head())

print("\nHourly Orders DataFrame Head:")
display(hourly_orders_df.head())

print("\nTop Pizzas DataFrame Head:")
display(top_pizzas_df.head())

print("\nPopular Size DataFrame Head:")
display(popular_size_df.head())
```

üìä **4. Visualizations**

üí∞ **Cumulative Revenue Over Time**
This plot shows the total revenue accumulated each day throughout the year.

**Business Insight:**
The cumulative revenue shows a steady upward trend throughout the year,
indicating consistent sales performance. The steeper slopes suggest periods of higher sales.

```python
import matplotlib.pyplot as plt

# Plot the cumulative revenue over time
plt.figure(figsize=(14, 6))
plt.plot(cumulative_revenue_df['date'], cumulative_revenue_df['cum_rev'], color='green', linewidth=2, label='Cumulative Revenue')

# Beautify the chart
plt.title('Cumulative Revenue Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

üïí **Hourly Order Volume**
This plot shows the number of orders placed in each hour of the day.

**Business Insight:**
Most customer orders are placed between 12 PM and 1 PM, showing a strong lunch-time trend.
This helps the business optimize staff shifts, prepare kitchen resources, and plan inventory more efficiently.

```python
import matplotlib.pyplot as plt

# Plot hourly orders volume
plt.figure(figsize=(8, 5))
plt.plot(hourly_orders_df['hr'], hourly_orders_df['hourly_orders'], marker='o', color='blue')
plt.title("Hourly Order Volume")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.xticks(hourly_orders_df['hr'])
plt.tight_layout()
plt.show()
```

üìä **Top 5 Most Ordered Pizzas**
This plot identifies the most popular pizzas by quantity ordered.

**Business Insight:**
The Classic Deluxe and Barbecue Chicken Pizzas lead sales volume,
indicating customer preference for hearty, flavorful options. These can be prioritized for promotions and stocking.

```python
import matplotlib.pyplot as plt

# Plot the top 5 most ordered pizzas by quantity
plt.figure(figsize=(8, 5))
plt.bar(top_pizzas_df['pizza_type'], top_pizzas_df['total_quantity'], color='orange')
plt.title("Top 5 Most Ordered Pizzas")
plt.xlabel("Pizza Type")
plt.ylabel("Quantity Ordered")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

üí∞ **Top 3 Pizzas by Revenue**
This plot shows the highest-earning pizzas by total revenue.

**Business Insight:**
The most popular pizzas by quantity are also the highest revenue generators.
Focusing marketing and inventory on these items is crucial.

```python
import matplotlib.pyplot as plt

# Plot the top 3 pizzas by revenue
plt.figure(figsize=(8, 5))
plt.bar(top_pizzas_revenue_df['pizza_type'], top_pizzas_revenue_df['total_revenue'], color='purple')
plt.title("Top 3 Pizzas by Revenue")
plt.xlabel("Pizza Type")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

üçï **Pizza Size Distribution (All Sizes Included)**
This plot shows the quantity ordered for each pizza size.

**Business Insight:**
Medium and Large pizzas dominate sales, indicating customer preference for standard portions.
Upsell larger sizes or bundle smaller ones for better margins.

```python
import matplotlib.pyplot as plt

# Plot the distribution of pizza sizes ordered
plt.figure(figsize=(7, 5))
plt.bar(popular_size_df['size'], popular_size_df['total_quantity'], color='tomato')
plt.title("Pizza Size Ordered")
plt.xlabel("Size")
plt.ylabel("Quantity Ordered")
plt.tight_layout()
plt.show()
```

üìä **Top 3 Ordered Pizzas by Category**
This plot breaks down the top-performing pizzas within each category by revenue.

**Business Insight:**
Understanding the top performers within each category allows for targeted marketing
and menu optimization.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the top 3 ordered pizzas by category based on revenue
plt.figure(figsize=(12, 7))
sns.barplot(x='name', y='cat_pizza_rev', hue='category', data=top_3_by_category_df)
plt.title("Top 3 Ordered Pizzas by Category (by Revenue)")
plt.xlabel("Pizza Name")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

üìä **Pizza Quantity by Category**
This plot shows the total orders for each pizza category.

**Business Insight:**
Understanding the distribution of orders across different pizza categories
helps in managing inventory and tailoring marketing efforts.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the total quantity ordered for each pizza category
plt.figure(figsize=(8, 5))
sns.barplot(x='category', y='category_quantity', data=category_quantity_df, color='skyblue')
plt.title("Pizza Quantity by Category")
plt.xlabel("Category")
plt.ylabel("Quantity Ordered")
plt.tight_layout()
plt.show()
```

üìä **Pizza Quantity Distribution by Category**
This plot shows the proportion of orders for each pizza category using a pie chart.

**Business Insight:**
Understanding the proportion of orders for each pizza category
helps in strategic planning, inventory management, and targeted marketing campaigns.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate the distribution percentage
category_quantity_df['distribution'] = (category_quantity_df['category_quantity'] / category_quantity_df['category_quantity'].sum()) * 100

# Plot the pie chart for pizza quantity distribution by category
plt.figure(figsize=(8, 8))
plt.pie(category_quantity_df['category_quantity'], labels=category_quantity_df['category'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Pizza Quantity Distribution by Category")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
